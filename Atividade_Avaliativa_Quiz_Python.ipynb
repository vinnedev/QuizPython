# Pergunta 2

A reforma tributária em Autonomópolis, o governo que presa pela efetividade na prestação dos serviços público, tendo metas, métricas, análises de resultados e resolutividade com foco na satisfação dos contribuintes, pretende atrair mais investidores, mais empreendedores e fomentar, ainda mais a economia inovadora, reduziu a tributação, fixando um imposto único de 10% sobre a renda maior que um salário mínimo de AU$ 1302.00 e, 0% sobre a produção e o patrimônio. Neste cenário, Autonomópolis cresce, em média, 11% ao ano, com inflação próxima de zero. Como a burocracia estatal é mínima, simplesmente operacional e funcional, os cálculos tributários são simples. O software a seguir permite o cálculo da folha de pagamentos das empresas com funcionários horistas:

image.png

FUNCIONÁRIO 1:
Salário hora: AU$ 10,00
Horas Trabalhadas: 120

FUNCIONÁRIO 2:
Salário hora: AU$ 15,00
Horas Trabalhadas: 100

````python
# Pergunta 2

print("InlnSiCaSa - Sistema para Cálculo de Salário\n")
NFuncionarios = int(input("Informar o número de funcionários: "))

TotalSalarioBruto = 0
TotalImposto = 0
TotalSalarioLiquido = 0

for i in range(NFuncionarios):
    print("\nFUNCIONÁRIO [", i, "]:")
    sh = float(input("DIGITE O SALÁRIO HORA DO FUNCIONÁRIO: "))
    ht = float(input("DIGITE O NÚMERO DE HORAS TRABALHADAS: "))
    sb = sh * ht

    VImposto = 0
    if sb > 1302:
        VImposto = sb * 10 / 100

    sl = sb - VImposto

    print("\nSALÁRIO BRUTO: ", sb)
    print("IMPOSTO: ", VImposto)
    print("SALÁRIO LÍQUIDO: ", sl)

    TotalSalarioBruto = TotalSalarioBruto + sb
    TotalImposto = TotalImposto + VImposto
    TotalSalarioLiquido = TotalSalarioLiquido + sl

print("\nTOTALIZAÇÃO: ")
print("TOTAL BRUTO DA FOHA DE PAGAMENTOS : ", TotalSalarioBruto)
print("ToTAL IMPOSTO ARRECADADO : ", TotalImposto)
print("ToTAL Líquido DA FOHA DE PAGAMENTOS: ", TotalSalarioLiquido)
````
